# 🎓 Student Performance Analysis & Prediction System

## 🎯 Objective

The objective of this project is to analyze student performance based on factors like study hours, attendance, and previous marks, and to predict final marks using a rule-based approach.

## 📊 Dataset Description

The dataset contains the following features:

- StudyHours: Number of hours a student studies
- Attendance: Percentage of attendance
- PreviousMarks: Marks obtained in previous exams
- FinalMarks: Final marks scored by the student

## 🧹 Data Cleaning

- Checked for missing values
- Verified data types
- Removed invalid or null values

## 📈 Data Visualization

- Scatter plot between Study Hours and Final Marks
- Scatter plot between Attendance and Final Marks
- Correlation Heatmap to analyze relationships

## 🔍 Data Analysis

- Students who study more tend to score higher marks
- Higher attendance leads to better performance
- Previous marks also influence final marks

## 🔮 Prediction System

A rule-based system is used to predict student performance using the formula:

Score = (StudyHours × 5) + (Attendance × 0.3) + (PreviousMarks × 0.2)

The system also assigns grades based on predicted marks.

## 🏁 Conclusion

This project demonstrates how data analysis and rule-based logic can be used to predict student performance without using machine learning algorithms.

```python
# ================== FIX DATA TYPES ==================
data['StudyHours'] = pd.to_numeric(data['StudyHours'], errors='coerce')
data['Attendance'] = pd.to_numeric(data['Attendance'], errors='coerce')
data['PreviousMarks'] = pd.to_numeric(data['PreviousMarks'], errors='coerce')
data['FinalMarks'] = pd.to_numeric(data['FinalMarks'], errors='coerce')
```

```python
# ================== DROP NULL ==================
data = data.dropna()
```

```python
import matplotlib.pyplot as plt

plt.scatter(data['StudyHours'], data['FinalMarks'])
plt.xlabel("Study Hours")
plt.ylabel("Final Marks")
plt.title("Study Hours vs Final Marks")
plt.show()
```

```python
plt.scatter(data['Attendance'], data['FinalMarks'])
plt.xlabel("Attendance")
plt.ylabel("Final Marks")
plt.title("Attendance vs Final Marks")
plt.show()
```
